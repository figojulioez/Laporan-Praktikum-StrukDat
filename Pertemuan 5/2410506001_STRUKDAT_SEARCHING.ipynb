```python
def  linear_search(arr, target) :
  # melakukan perulangan sebanyak panjang dari array tersebut
  for i in range(len(arr)) :
    # mencari target dari perulangan satu persatu
    if arr[i] == target :
      # Mengembalikan nilai tersebut
      return i
    # Jika tidak ditemukan mengembalikan nilai -1
    return -1

# Mengisi dengan angka acak dan NPM saya dan mengurutkannya
arr = sorted([24, 50, 10, 40, 1, 60])


# Mengembalikan target
target = int(input("Masukan 3 digit terakhir NPM Anda: "))

# menyimpan hasil
result = linear_search(arr, target)

if result != -1:
  print(f"Elemen ditemukan pada indeks {result}")
else :
  print("Elemen tidak ditemukan")
```

```python
def binary_search (arr, target):
  # Menentukan batas bawah
  low = 0

  # Menentukan batas atas
  high = len(arr) - 1

  # Perulangan akan terjadi selama nilai low lebih kecil sama dengan dengan nilai tengah
  while low <= high:
      # Mencari nilai tengah untuk pencarian
      mid = (low + high) // 2

      # Jika nilai tengah nilainya sama dengan target
      if arr[mid] == target:
        # Maka akan mengembalikan nilai itu sendiri
        return mid

      # Jika nilai tengah lebih kecil dari target maka pencarian akan berlanjut ke sebelah kanan
      elif arr[mid] < target:
        # memajukan indeks low
        low = mid + 1

      # Asumsi jika nilai tengah lebih besar dari target maka pencarian akan berlanjut ke sebelah kiri
      else:
        high = mid - 1
  return -1

# Data terurut dengan tanggal lahir Anda (DDMM)
arr = sorted([707, 1210, 2807, 1307, 2012, 3009, 2802, 101])

# Meminta nilai input
target = int(input("Masukkan tanggal lahir Anda (DDMM): "))

result = binary_search(arr, target)
  # Jika nilainya ditemukan
if result != -1:
  print(f"Elemen ditemukan pada indeks {result}")
else:
  print("Elemen tidak ditemukan")
```

```python
# mengimport library math
import math

# mendefinisikan fungsi jump search
def jump_search(arr, target) :
  # panjang dari array
  n = len(arr)
  # digunakan sebagai batas dari pencarian linear nantinya
  # akar kuadrat dari panjang array tersebut
  step = int(math.sqrt(n))
  # digunakan untuk awal dari pencarian linear nantinya
  prev = 0

  # Jika kondisi arr dengan indeks nilai minimum antara step dan panjang array dikurang 1 lebih kecil dari target
  while arr[min(step, n) - 1] < target :
    # batas awal akan dimajukan sebesar step tersebut
    prev = step
    # lalu batas akhirnya akan dimajukan juga sebanyak jumlah akar kuadrat dari panjang array tersebut
    step += step

    # Jika setelah melakukan pencarian ternyata batas bawah nilainya lebih besar atau sama dengan n
    # maka target tidak ditemukan
    if prev >= n :
      return -1


  # melakukan pencarian secara linear jika sudah ditemukan batas atas dan batas bawahnya
  # melakukan perulangan dari batas bawah sampai batas atas
  for i in range(prev, min(step, n)):

    # jika target ditemukan maka
    if (arr[i] == target) :
      # mengembalikan nilai dari target dan jumlah perulangan
      return i
  return -1

# Data bedasarkan nama teman - teman saya yang sudah di urutkan
arr = sorted(["Zico", "Arya", "Aila", "Ica", "Abdi", "Rizal", "Rayfal", "David"])
target = input("Masukkan inisial nama Anda: ")

# menghasilkan nilai
result = jump_search(arr, target)

if result != -1:
  print(f"Elemen ditemukan pada indeks {result}")
else :
  print("Elemen tidak ditemukan")


```

# LINEAR SEARCH

```python
# membuat fungsi linear search dengan parameter array dan targer
def linearSearch (arr, target) :
  # membuat sebuah counter untuk menghitung berapa banyak waktu perulangan yang  dibutuhkan
  counter = 0

  # Membuat perulangan  secara linear dengan jumlah perulangan adalah panjang array itu sendiri
  for i in range(len(arr)) :

    # Menambah counter dengan 1 setiap kali perulangan terjadi
    counter += 1

    # Jika arr nilainya sama dengan dari dari target
    if (arr[i] == target) :
      # Mengembalikan nilai dari target dan jumlah perulangan
      return [arr[i], counter]

  return [-1, counter]
```

# Binary search

```python
# membuat fungsi binary search dengan paramater arr dan targert
def binarySearch(arr, target) :
  # mendefinisikan indeks terkecil
  low = 0
  # mendefinisikan indeks terbesar
  high = len(arr) - 1
  # Mendifinisikan counter untuk mengetahui jumlah perulangan yang terjadi
  counter = 0

  # Kondisi apa bila low nilainya lebih kecil dari high
  while low <= high :
    # Menambahkan nilai counter setiap kali perulangan
    counter += 1

    # Mengambil indeks nilai tengah dengan rumus penjumlahan dari low dan high lalu dibagi bulat dengan 2
    mid = (low + high) // 2
    # cek apakah array indeks ke mid nilainya sama dengan target
    if arr[mid] == target :
      # jika nilainya sama maka akan mengembalikan nilai dari target dan jumlah perulangan
      return [arr[mid], counter]

    # Jika nilainya lebih kecil dari target maka indeks low akan menjadi indeks ke mid + 1
    elif arr[mid] < target :
      # memajukan indeks dari low tersebut
      low = mid + 1

    # Jika arr[mid] nilainya lebih besar dari target maka indeks high akan menajadi indeks ke mid - 1
    else :
      # memundurkan indeks dari high tersebut
      high = mid - 1

  # Jika selama perulangan tidak ada nilai yang ditemukan
  return [-1, counter]
```

# Jump Search

```python
# mengimport library math
import math

# mendefinisikan fungsi jump search
def jumpSearch(arr, target) :
  # panjang dari array
  n = len(arr)
  # digunakan sebagai batas dari pencarian linear nantinya
  # akar kuadrat dari panjang array tersebut
  step = int(math.sqrt(n))
  # digunakan untuk awal dari pencarian linear nantinya
  prev = 0

  # Membuat sebuah counter untuk menghitung jumlah perulangan yang terjadi
  counter = 0

  # Jika kondisi arr dengan indeks nilai minimum antara step dan panjang array dikurang 1 lebih kecil dari target
  while arr[min(step, n) - 1] < target :
    # menghitung jumlah counter
    counter += 1
    # batas awal akan dimajukan sebesar step tersebut
    prev = step
    # lalu batas akhirnya akan dimajukan juga sebanyak jumlah akar kuadrat dari panjang array tersebut
    step += step

    # Jika setelah melakukan pencarian ternyata batas bawah nilainya lebih besar atau sama dengan n
    # maka target tidak ditemukan
    if prev >= n :
      return [-1, counter]

  # melakukan pencarian secara linear jika sudah ditemukan batas atas dan batas bawahnya
  # melakukan perulangan dari batas bawah sampai batas atas
  for i in range(prev, min(step, n)):
    # pada jumpsearch saya mengasumsikan setiap ada perulangan maka akan selalu +1
    counter += 1

    # jika target ditemukan maka
    if (arr[i] == target) :
      # mengembalikan nilai dari target dan jumlah perulangan
      return [arr[i], counter]

  return [-1, counter]

```
