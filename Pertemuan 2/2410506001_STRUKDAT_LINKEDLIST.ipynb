```python
# Definisi Node dalam Linked List
class TaskNode :
  def __init__ (self, task_name) :
    self.task_name = task_name
    self.next = None

class TaskList :
  def __init__ (self) :
    self.head = None

  def add_task (self, task_name) :
    new_task = TaskNode(task_name)
    if not self.head :
      self.head = new_task

    else :
      current = self.head
      while current.next :
        current = current.next
      current.next = new_task

  def show_tasks(self) :
    if not self.head:
      print("\n Tidak ada tugas hari ini!")
      return

    print("\n Daftar Tugas Hari Ini:")
    current = self.head
    count = 1
    while current :
      print(f"{count}. {current.task_name}")
      current = current.next
      count += 1

task_list = TaskList()

num_tasks = int(input("Masukkan jumlah tugas hari ini: "))

for i in range(1, num_tasks + 1):
    task_name = input(f"Masukkan nama tugas ke-{i}: ")
    task_list.add_task(task_name)

task_list.show_tasks()
```

```python
# Definisi Node dalam Linked List
class mataKuliahNode :
# Constructor method pada mataKuliahNode
  def __init__ (self, mataKuliah_name) :
  # Menyimpan atribut mataKuliah_name yang di isi argument mataKuliah_name
    self.mataKuliah_name = mataKuliah_name
  # Digunakan untuk referensi data selanjutnya
    self.next = None

# Function Spasi Agar tulisan terlihat lebih rapih
# Lebih detainya ada pada https://medium.com/@figojulioez/python-operator-pada-python-6556e0fc6210
def spasi(awal, akhir) :
  total = 50;
  spasi = total - len(awal) - 2
  print(awal, " " * spasi,":", akhir)
  print("")

# Membuat Sebuah Class Linked List
class mataKuliahList :
# Constructor pada Class mataKuliahList
  def __init__ (self) :
# Pada saat awal inisiasi class data self.head kosong
    self.head = None

# Method menambahkan matakuliah
  def add_mataKuliah (self, mataKuliah_name) :
  # Inisiasi cikal bakal node baru
    new_mataKuliah = mataKuliahNode(mataKuliah_name)
    # Apabila self.head masih kosong
    if not self.head :
      # Maka self.head mereferensi new_matakuliah
      self.head = new_mataKuliah

    # Jika Kondisi Awal Tidak Terpenuhi
    else :
      # current mereferensi ke self.head
      current = self.head
      # Akan melooping hingga current.next == none
      while current.next :
        # Memindahkan pointer jadi current bereferensi ke objek selanjutnya
        current = current.next
      # Menambahkan new_matakuliah pada current.next
      current.next = new_mataKuliah

  # Method untuk menampilkan matakuliah
  def show_mataKuliahs(self) :
  # Jika tidak ada data yang ditampilkan
    if not self.head:
      print("\n Tidak ada mata kuliah hari ini!")
      return

  # Asumsi jika ada data yang ditampilkan
    print("\n Daftar mata kuliah Ini:")
    # current mereferensi objek self.head
    current = self.head
    # Untuk Perulangan nantinya
    count = 1
    # Looping yang akan berlanjut sampai datanya kosong
    while current :
      # Menampilkan output dengan function spasi agar lebih rapih
      spasi(f"{count}. {current.mataKuliah_name['mataKuliah']}", f"{current.mataKuliah_name['jamMulai']} - {current.mataKuliah_name['jamSelesai']}")
      # Setelah itu maka akan berpindah node atau berpindah referensi
      current = current.next
      # Menjumlahkan count untuk penomoran berikutnya
      count += 1
# Menginisiasi class mataKuliahList
mataKuliah_list = mataKuliahList()

# Meminta pengguna ingin menginpukan berapa mata kuliah
num_mataKuliahs = int(input("Masukkan jumlah mata kuliah hari ini: "))

# Melakukan looping sesuai yang diinginkan pengguna
for i in range(1, num_mataKuliahs + 1):
    # Meminta input nama mata kuliah
    mataKuliah = input(f"Masukkan nama mata kuliah ke-{i}: ")
    # Meminta input jam mulai dan jam selesai
    jamMulai  = str(input(f"Masukkan jam mulai mata kuliah ke-{i}: "))
    jamSelesai = str (input(f"Masukkan jam selesai mata kuliah ke-{i}: "))
    # Mengirimkan data ke metjod add dalam bentuk dictionary
    mataKuliah_list.add_mataKuliah({"mataKuliah" : mataKuliah, "jamMulai" : jamMulai, "jamSelesai" : jamSelesai})

# Menampikan mata kuliah
mataKuliah_list.show_mataKuliahs()
```

```python
class PageNode :
  def __init__ (self, url) :
    self.url = url
    self.next = None
    self.prev = None

class BrowserHistory :
  def __init__(self) :
    self.current = None

  def visit_page (self, url) :
    new_page = PageNode(url)
    if not self.current :
      self.current = new_page

    else :
      new_page.prev = self.current
      self.current.next = new_page
      self.current = new_page

  def show_history (self) :
    if not self.current :
      print("\n Riwayat kosong. ")
      return

    print("\n Riwayat Peramban Web:")
    temp = self.current
    while temp.prev:
      temp = temp.prev

    count = 1
    while temp :
      print(f"{count}. {temp.url}")
      temp = temp.next
      count += 1

  def go_back(self) :
      if self.current and self.current.prev :
        self.current = self.current.prev
        print(f"\n Anda kembali ke: {self.current.url}")
      else :
        print(f"\n Tidak bisa kembali, ini adalah halaman pertama!")

  def go_forward(self) :
      if self.current and self.current.next :
        self.current = self.current.next
        print(f"\n Anda maju ke: {self.current.url}")
      else :
        print(f"\n Tidak bisa maju, ini adalah halaman terakhir!")


history = BrowserHistory()

num_pages = int(input("Masukkan jumlah halaman yang ingin dikunjungi: "))

for i in range(1, num_pages + 1):
    url = input(f"Masukkan URL halaman ke-{i}: ")
    history.visit_page(url)

history.show_history()

while True :
  command = input("\n Ketik 'b' untuk kembali, 'f' untuk maju, tekan 'x' untuk keluar.\nMasukan perintah: ").lower()

  if command == 'b' :
    history.go_back()
  elif command == 'f' :
    history.go_forward()
  elif command == 'x' :
    break
  else :
    print("\n Perintah tidak dikenal.")
```

# Latihan 2
Buatlah program yang menampilkan riwayat navigasi folder dalam File Explorer menggunakan Dobly Linked List. Jumlah folder yang anda navigasikan sesuai dengan digit terakhir NPM anda!

```python
# Class FolderNode yang menjadi struktur node dari folder - folder
class FolderNode :
# Constructor method
  def __init__ (self, folder) :
    # Untuk Menyimpan nama dari folder
    self.folder = folder
    # Mereferensi data self.next dan self.prev
    self.next = None
    self.prev = None

# Inisiasi Class FolderHistory
class FolderHistory :
  # Constructor method
  def __init__(self) :
    # Di awal inisiasi objek atribut current kosong
    self.current = None

  # Method untuk membuat folder baru
  def visit_folder (self, folder) :
    # Instansi Objek FolderNode
    new_folder = FolderNode(folder)
    # Jika atribut current kosong di awal
    if not self.current :
      # Maka atribut current akan mereferensi ke new_folder
      self.current = new_folder

    # Jika tidak
    else :
      # new_folder.prev mereferensi ke self.current yang artinya prev <- current
      new_folder.prev = self.current
      # self.current.next mereferensi ke prev maka current -><- prev saling terhubung
      self.current.next = new_folder
      # Untuk pindah jalur yang dimana data lama akan di ganti data baru jalurnya
      self.current = new_folder

  # Method untuk menampilkan history
  def show_history (self) :
    # Jika atribut current kosong
    if not self.current :
      print("\n Riwayat kosong. ")
      return

    # Asumsi atribut current memiliki nilai
    print("\n Riwayat Peramban Folder:")
    # temp mereferensi ke objek self.current
    temp = self.current
    # Akan mengulang sampai temp.prev nya kosong
    while temp.prev:
      # mereferensi ke data sebelumnya karena jalur pada current yang baru
      temp = temp.prev


    count = 1
    # Mengulang temp sampai datanya kosong
    while temp :
      print(f"{count}. {temp.folder}")
      # temp mereferensi ke node selanjuutnya
      temp = temp.next
      count += 1

# Method untuk kembali dari node selanjutnya
  def go_back(self) :
      # Jika jalur sebelumnya dengan jalur sekaang tidak kosong nilainya
      if self.current and self.current.prev :
        # Maka pinda jalurnya ke jalur sebelumnya
        self.current = self.current.prev
        print(f"\n Anda kembali ke: {self.current.folder}")
      # Jika ternyata jalurnya sama
      else :
        print(f"\n Tidak bisa kembali, ini adalah folder pertama!")

# Method untuk lanjut dari node sebelumnya
  def go_forward(self) :
    # Jika jalur selanjutnya dan jalur sekarang tidak kosong nilainya
      if self.current and self.current.next :
        # current mereferensi ke current.next selanjutnya
        self.current = self.current.next
        print(f"\n Anda maju ke: {self.current.folder}")
         # Jika tidak
      else :
        print(f"\n Tidak bisa maju, ini adalah folder terakhir!")

# Instansi Objek FolderHistory
history = FolderHistory()

# Meminta jumlah folder ke pengguna
num_folder = int(input("Masukkan jumlah folder yang ingin dikunjungi: "))

# Looping sebanyak jumlah pengguna
for i in range(1, num_folder + 1):
    # Untuk meminta nama faolder yang di inginkan
    folder = input(f"Masukkan folder halaman ke-{i}: ")
    # Lalu menambahnya dengan method .visit_folder()
    history.visit_folder(folder)
# Menampilkan Folder
history.show_history()

# Looping akan terus berulang
while True :
  # Meminta input untuk comandnya
  command = input("\n Ketik 'b' untuk kembali, 'f' untuk maju, tekan 'x' untuk keluar.\nMasukan perintah: ").lower()

# Jika command == b maka jalur node akan kembali ke current.prev
  if command == 'b' :
    history.go_back()
 # Jika command == b maka jaru node akan lanjut ke current.next
  elif command == 'f' :
    history.go_forward()
  # Keluar dari program
  elif command == 'x' :
    # Keluar dari perulangan
    break
  else :
    print("\n Perintah tidak dikenal.")
```

```python

class Node:
  # Akan dijalankan pertama kali saat class di instansi (Constructor)
  def __init__(self, data) :
    self.data = data # Menyimpan nilai dari node
    self.next = None # Pointer atau ekor ke node berikutnya

class LinkedList:
  # Akan dijalankan pertama kali saat class di instansi (Constructor)
  def __init__(self) :
    self.head = None # Digunakan untuk menyimpan data singly linked listnya nanti

  # Method untuk menambahkan data baru di awal
  def add_at_begining(self, data) :
    # Akan menginstansu class Node lalu di simpan di variabel new_node
    new_node = Node(data)
    # new_node.next akan merefrensi data lama atau self.head
    new_node.next = self.head
    # lalu data self.head akan di ganti dengan data new_node yang mana atribut next pada new_node mereferensi ke node yang lama
    # sederhananya setiap data baru akan menjadi kepala dan data lama menjadi ekor
    self.head = new_node

  # Method untuk menambahkan data baru di akhir
  def add_at_end(self, data) :
    new_node = Node(data)
    # Apabila self.head data pada awalnya kosong
    if self.head == None :
      # Maka self.head akan langsung di isi
      self.head = Node(data)

     # Jika tidak
    else :
      # temp akan mereferensi self.head untuk semntara
      temp = self.head
      # lalu akan melooping terus hingga mencapai data terakhir
      while temp.next :
        # Setiap berpindah node maka node tersebut akan di simpan di variabel next
        temp = temp.next
      # maka data baru tersebut di simpan di akhir
      temp.next = new_node

  # Method untuk menambahkan elemen di posisi tertentu
  def add_at_position(self, data, position) :
    # Jika posisi yang di inputkan < 0
    if position < 0:
      print(" Posisi tidak valid.")
      return

    # Menginisiasi class Node disimpan di new_node
    new_node = Node(data)
    # Jika posisi yang di minta adalah 0
    if position == 0:
      # Maka jalankan method ini, karena fungsinya menambah data di awal node
      self.add_at_begining(data)
      return

    # Jika kedua kondisi di atas tidak terpenuhi maka asumsinya angka yang di masukan lebih dari 0
    # temp mereferensi self.head
    temp = self.head
    # Melooping hingga mencapai posisi yang di inginkan
    for _ in range(position - 1) :
      # Jika posisi kosong atau none maka posisi terlalu panjang
      if not temp :
        print("Posisi melebihi panjang daftar!")
        return
      # temp mereferensi ke temp.next tapi perubahan referensi tersebut terhubung ke self.head
      temp = temp.next
    # Menyimpan referensi temp.next pada new_node.next
    new_node.next = temp.next
    # lalu merubah referensi tersebut temp.next menjadi new_node
    temp.next = new_node

  # Method untuk menghapus elemen di awalh
  def delete_at_begining(self) :
    # jika datanya kosong
    if not self.head :
      print("Daftar kosong, tidak ada yang bisa dihapus!")
    # maka referensi self.head di ubah ke self.head.next selanjutnya
    self.head = self.head.next

  # Method untuk menghapus elemen di akhir
  def delete_at_end(self) :
    # Jika self.headnya kosong nilainya
    if not self.head :
      print(" Daftar kosong, tidak ada yang bisa dihapus!")
      return
    # Jika self.head.next tidak memiliki data yang artinya datanya 1
    if not self.head.next :
      # Maka akan menjadikan self.head = none
      self.head = None

    # temp mereferensi ke self.head
    temp = self.head
    # Mengulang sampai data selanjutnya dari temp.next.next itu none
    while temp.next.next :
      # Mereferensi ke objek temp.next atau objek selanjutnya
      temp = temp.next
    # Membuat temp.next tersebut mejadi kosong
    temp.next = None

  # Method untuk menghapus posisi tertentu
  def delete_at_position(self, position) :
    # Jika data kosong
    if not self.head:
      print("Daftar kosong, tidak ada yang bisa dihapus!")
      return

    # Jika posisi yang di inputkan = 0
    if position == 0:
      # mengganti kepala dengan data selanjutnya
      self.head = self.head.next
      return
    # Mereferensi temp = self.head
    temp = self.head
    # perulangan utnuk menentukan posisi yang di hapus
    for _ in range(position - 1) :
      # Jika temp kosong atau temp.next tidak ada
      if not temp or not temp.next :
        print("Posisi tidak ditemukan!")
        return
      # variabel temp mereferensi ke temp.next selanjutnya
      temp = temp.next
    # mengganti temp.next dengan data selanjutnya yaitu temp.next.next
    temp.next = temp.next.next

  # Method untuk mencari elemen
  def search (self, key) :
    # temp mereferensi ke self.head
    temp = self.head
    # Melakukan perulangan sampai temp tidak kosong
    while temp :
      # Jika data yang dicari ditemukan maka akan mengembalikan nilai
      if (temp.data == key) :
        print(f" Elemen {key } ditemukan dalam daftar!")
        return
      # Jika datanya belum ditemukan maka akan mereferensi ke temp selanjutnya
      temp = temp.next
    # Jika selama perulangan data tidak ditemukan
    print(f"Elemen {key} tidak ditemukan dalam daftar.")

  # Method untuk membalik linked list
  def reverse(self) :
    # Data yang akan menyimpan keterbalikannya
    prev = None
    # current mereferensi objek self.head
    current = self.head
    # looping sampai current = None
    # Semisal ada data a => b => c
    while current :
      # next node akan menyimpan data b => c saja
      next_node = current.next
      # lalu akan menjadi seperti a => none
      current.next = prev
      # Dan prev akan menyimpan a => none
      prev = current
      # dan curren akan menyimpan b => c konsepnya hampir mirip dengan add_at_begining untuk di prev
      current = next_node
   # merubah referensi self.head = prev
    self.head = prev

  # method untuk menampilkan linked list
  def display(self):
    # jika self.head kosong atau datanya kosong
    if not self.head:
      print(" Daftar kosong!")
      return
    # temp mereferensi ke self.head
    temp = self.head
    # looping temp hingga datanya kosong
    while temp:
        print(temp.data, end=" -> ")
        # mereferensi temp selanjutnya ini mirip seperti i++
        temp = temp.next
    print("None")

# Membuat objek linked llist
ll = LinkedList()

# meminta pengguna memasukan elemen awal
num_elements = int(input("\nMasukan jumlah awal elemen dalam daftar: "))
# Melakukan perulangan sebanyak yang di inginkan pengguna
for i in range(num_elements) :
  # Menginput data yang ada
  data = int(input(f"Masukan data ke-{i + 1}"))
  # Menambahkan data di akhir
  ll.add_at_end(data)

# Menu Operasi Linked List
while True:
  print("\n Pilih Operasi:")
  print("1.   Tambah elemen di awal")
  print("2.   Tambah elemen di akhir")
  print("3.   Tambah elemen di posisi tertentu")
  print("4.   Hapus elemen di awal")
  print("5.   Hapus elemen di akhir")
  print("6.   Hapus elemen di posisi tertentu")
  print("7.   Cari elemen dalam daftar")
  print("8.   Balikan daftar")
  print("9    Tampilkan daftar")
  print("0    Keluar")

  choice = int(input("\nMasukkan pilihan: "))

  if choice == 1:
      data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
      ll.add_at_begining(data)
  elif choice == 2:
      data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
      ll.add_at_end(data)
  elif choice == 3:
      position = int(input("Masukkan posisi untuk menambahkan data: "))
      data = int(input("Masukkan data yang ingin ditambahkan: "))
      ll.add_at_position(position, data)
  elif choice == 4:
      ll.delete_at_begining()
  elif choice == 5:
      ll.delete_at_end()
  elif choice == 6:
      position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
      ll.delete_at_position(position)
  elif choice == 7:
      key = int(input("Masukkan elemen yang ingin dicari: "))
      ll.search(key)
  elif choice == 8:
      ll.reverse()
      print(" Daftar berhasil dibalik!")
  elif choice == 9:
      ll.display()
  elif choice == 0:
      print("\n Keluar dari program.")
      break
  else:
      print("\n Pilihan tidak valid, silakan coba lagi!")








```
