# Tower of Hanoi

```python
# Menggunakan library random untuk mengacak nomor
import random as rand
from math import ceil
# Mendefinisikan class Tower Of Hanoi
# Membuat class tower of hanoi
class TowerOfHanoi:
  # Method constructor
  def __init__(self) :
    # Membuat variabel tiang a,b,c untuk menyimpan cakram yang berupa bintang
    self.tiangA = []
    self.tiangB = []
    self.tiangC = []
    # Untuk jumlah cakram, jumlah cakram ini akan di kurang 1. jadi jika 5 maka jumlah cakramnya akan 4
    self.jumlahCakram = 5

  # Membuat method untuk mengacak cakram tersebut
  def acakCakram(self) :
    # Tiang yang nantinya akan digunakan dalam rand
    tiang = ["a", "b", "c"]
    # memasukan cakram ke tiang yang sudah ditentukan
    for i in range(2, self.jumlahCakram  * 2, 2) :
      # Memilih ingin mengacak tiang secara random
      acakTiang = rand.randint(1,3)

      # Jika Tiang == 1 maka akan memasukan cakram ke tiang A
      if acakTiang == 1 :
        self.push("a", i)

      # Jika Tiang == 2 maka akan memasukan cakram ke tiang B
      if acakTiang == 2 :
        self.push("b", i)

      # Jika Tiang == 2 maka akan memasukan cakram ke tiang C
      if acakTiang == 3 :
        self.push("c", i)

  # Method tampilkan cakram
  def tampilkanCakram (self) :
    # Mencari panjang maximum untuk nantinya akan digunakan untuk membuat tiang
    print("Tampilkan jumlah cakram ", 'A : ', self.tiangA, '|| B : ', self.tiangB, '|| C : ', self.tiangC)
    print("")
    # Akan mengulang semua tiang dan mengisi datanya
    for i in range(self.jumlahCakram - 2, -1, -1) :
      # Jarak untuk setiap baris yang akan kita gunakan untuk jarak tiang
      spasi = 0
      # Disini saya menggunakan try catch untuk antisipasi kelebihan
      # Membuat Tiang A
      try :
        # Menghitung spasi supaya tampilan cakram lebih rapih dengan rumus jumlah dari cakram dikalikan 2 lalu dikurang nilai tiang dan dibagi 2
        # contoh : jika nilai tiangA = 2, maka bintang tersebut ada di posisi 10 dan 11
        # Maka : pembulatan dari 10 * 2 - nilaiTiangA /2 = 9
        spasi += ceil((self.jumlahCakram * 2 - self.tiangA[i]) / 2 )
        print(f"{' ' *  spasi +  '*' * self.tiangA[i]}", end="")
        spasi = ceil((self.jumlahCakram * 2 - self.tiangA[i]) / 2 )
      except IndexError:
        # Jika cakram nya kosong maka saya asumsikan dia ada di setengah nilai dari jumlahCakramnya
        spasi += ceil((self.jumlahCakram * 2 - 2) / 2 )
        print(f"{' ' * spasi +  '|'}", end="")
        # Karena dia hanya menampilkan 1 huruf maka saya tambahkan 1 karena setiap baris memiliki minimal 2 hruf
        spasi = ceil((self.jumlahCakram * 2 - 2) / 2 ) + 1


      # Membuat Tiang B
      try :
        # Menghitung spasi supaya tampilan cakram lebih rapih dengan rumus jumlah dari cakram dikalikan 2 lalu dikurang nilai tiang dan dibagi 2
        # contoh : jika nilai tiangA = 2, maka bintang tersebut ada di posisi 10 dan 11
        # Maka : pembulatan dari 10 * 2 - nilaiTiangA /2 = 9
        spasi += ceil((self.jumlahCakram * 2 - self.tiangB[i]) / 2 )
        print(f"{' ' * spasi +  '*' * self.tiangB[i]}", end="")
        spasi = ceil((self.jumlahCakram * 2 - self.tiangB[i]) / 2 )
      except IndexError:
        spasi += ceil((self.jumlahCakram * 2 - 2) / 2 )
        print(f"{' ' * spasi +  '|'}", end="")
        spasi = ceil((self.jumlahCakram * 2 - 2) / 2 ) + 1

      try :
        # Menghitung spasi supaya tampilan cakram lebih rapih dengan rumus jumlah dari cakram dikalikan 2 lalu dikurang nilai tiang dan dibagi 2
        # contoh : jika nilai tiangA = 2, maka bintang tersebut ada di posisi 10 dan 11
        # Maka : pembulatan dari 10 * 2 - nilaiTiangA /2 = 9
        spasi += ceil((self.jumlahCakram * 2 - self.tiangC[i]) / 2 )
        print(f"{' ' * spasi +  '*' * self.tiangC[i]}", end="")
        spasi = ceil((self.jumlahCakram * 2 - self.tiangC[i]) / 2 )
      except IndexError:
        spasi += ceil((self.jumlahCakram * 2 - 2) / 2 )
        print(f"{' ' * spasi +  '|'}", end="")
        spasi = ceil((self.jumlahCakram * 2 - 2) / 2 ) + 1

      print("")
    # Menampilkan keterangan cakram
    spasi = (ceil((self.jumlahCakram * 2 - 8) / 2))
    print(" " * spasi, "Tiang A", " " * (spasi * 2 - 1),"Tiang B", " " * (spasi * 2 - 1), "Tiang C" )

  # Method untuk menambahkan cakram ke tiang tertentu
  def push (self, tiang, data) :
    # Memilih tiang yang akan di tambahkan
      if (data == 'x') : return 0

      if tiang.lower() == "a" : self.tiangA.append(data)
      if tiang.lower() == "b" : self.tiangB.append(data)
      if tiang.lower() == "c" : self.tiangC.append(data)

  # Method untu hapus elemen terakhir
  def pop (self, tiang) :
    # gunakan try except jika datanya kosong
    try :
      if tiang.lower() == "a" : return self.tiangA.pop()
      if tiang.lower() == "b" : return self.tiangB.pop()
      if tiang.lower() == "c" : return self.tiangC.pop()
    except IndexError :
      print(f"Tiang {tiang.upper()} sudah kosong!")
      return "x"


tower = TowerOfHanoi()

# Memulai Game
print("Selamat Datang di Game Tower of Hanoi")
print("Urutkan cakram dari yang terbesar hingga terkecil")
print("Memulai acak cakram ...")
print("")

# Mengacak cakram
tower.acakCakram()
# membuat variabel jika kondisinya terpenuhi maka menang
menang = []
menang.extend(tower.tiangA)
menang.extend(tower.tiangB)
menang.extend(tower.tiangC)


menang = sorted(menang, reverse=True)

while True :
  # Menampilkan cakram
  tower.tampilkanCakram()
  print("=" * 60)
  print("")
  # Membuat pilihan untuk memindahkan cakram dari tiang apa
  pilihan = str(input("Pilih tiang yang ingin di ambil => ")).lower()
  # Membuat pilihan untuk memindahkan cakram ke cakram apa
  pindahkan = str(input("Pindahkan ke cakram ke tiang apa => ")).lower()
  # kondisi yang tidak sesuai
  if (pilihan not in ['a', 'b', 'c']) : print("Pilihan tidak sesuai");continue
  if (pindahkan not in ['a', 'b', 'c']) : print("Pilihan tidak sesuai");continue

  # pindahkan cakram
  tower.push(pindahkan,tower.pop(pilihan))

  # Jika tiangnya sama nilainya dengan menang maka dia menang
  if (tower.tiangA == menang or tower.tiangB == menang or tower.tiangC == menang ) :
    tower.tampilkanCakram()
    print("Selamat anda menang")
  else : continue

  break
```
