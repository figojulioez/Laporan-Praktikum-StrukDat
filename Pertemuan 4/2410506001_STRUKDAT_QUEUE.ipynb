```python
# Membuat class Queue dengan nama AntreanBioskop
class AntreanBioskop :
  # Constructor Method
  def __init__(self) :
    self.antrian = []

  # Membuat method yang mana pelanggan vip tidak akan mengantri
  def vipCustomer(self, kodeAntrian) :
    # Jika kode antrian berisi angka 7 (kode khusus) maka dapat mendapatkan kursi tanpa harus mengantri
    # mengembalikan nilai true jika berisi kode khusus
    return '7' in kodeAntrian

  # Method untuk menambahkan pelanggan ke dalam antrian
  def enqueue(self, ticket_id) :

    # Jika kode antrian berisi kode khusus
    if self.vipCustomer(ticket_id) :
      print("Selamat kode antrian anda berisi kode khusus")
      self.antrian.insert(0, ticket_id)
      print(f"Kode antrian {self.dequeue()} berhasil di proses")
      return 0

    # Jika antriannya penuh maka harus menunggu hinga slot kosong batas antrian 10
    if (self.size() >= 10) :
      print("Slot antrian penuh, tunggu hingga slot kosong.");
      return 0

    # Menambahkan ticket_id ke self.antrian
    self.antrian.append(ticket_id)
    # Menampilkan pesan bahwa antrian berhasil ditambahkan
    print(f"Kode Antrian : {ticket_id} berhasil di tambahkan kedalam antrian")

  # Method untuk mengembalikan panjang dari antrian
  def size(self) :
    # Mengembalikan panjang dari nilai self.antrian
    return len(self.antrian)

  # Memproses antrian dan mengeluarkannya dari antrian
  def dequeue(self) :
    # Jika datanya kosong
    if self.is_empty() : return None

    # Asumsi datanya ada
    return self.antrian.pop(0)

  # Method untuk cek apakah datanya kosong atau tidak
  def is_empty(self) :
    # Mengembalikan true jika datanya kosong
    return len(self.antrian) == 0

  # Method untuk menampilkan elemen paling depan
  def front(self) :
    # Operator Kondisional
    # Jika datanya kosong maka mengembalika none jika ada datanya mengembalikan nilai antrean dengan indeks pertama
    return None if self.is_empty() else self.antrian[0]

  # Method untuk menampilkan data paling akhir
  def rear(self) :
    # Operator Kondisional
    # Jika datanya kosong maka mengembalika none jika ada datanya mengembalikan nilai antrean dengan indeks terakhir
    return None if self.is_empty() else self.antrian[-1]

# Instansi class AntreanBioskop
antrian = AntreanBioskop()

# Nama 2 huruf pertama mahasiswa
namaAwal = ["JU", "NA", "KA", "AP", "DA", "MU", "FE", "RA", "RA", "FA", "SE", "NA", "SE", "ID", "EV", "DZ"]

# Untuk menambahkan data sebagai testing
for i in range(0, len(namaAwal)) :
  # Menambahkan kode Antrian
  antrian.enqueue(f"{namaAwal[i]}{f'0{i + 1}' if i in range(0,9) else i + 1}")



while True :
  print("")
  print("Selamat Datang di Antrian Bioskop")
  print(f"Antrian saat ini : {antrian.antrian}")
  print("1. Menambahkan antrian.")
  print("2. Memproses antrian.")
  print("3. Kode antrian pertama.")
  print("4. Kode antrian terakhir.")
  print("5. Panjang antrian.")
  print("6. Keluar.")

  print("")
  # Input menu yang di pilih user
  menu = int(input("Pilihlah salah satu dari menu tersebut => "))
  print("")
  # Menu pilihan
  # Jika menu == 1 maka menambahkan kode antrian
  if menu == 1 :
    antrian.enqueue(f"{str(input('Masukan kode antrian => '))}")
  # lain jika 2 memproses kode antrian
  elif menu == 2:
    print("Memproses:", antrian.dequeue())
  # lain jika 3 menampilkan kode antrian pertama
  elif menu == 3 :
    print("Kode antrian pertama:", antrian.front())
  # lain jika 4 menampilkan kode antrian terakhir
  elif menu == 4 :
    print("Rear element:", antrian.rear())
  # lain jika 5 menampilkan panjang antrian
  elif menu == 5 :
    print("Panjang antrian:", antrian.size())
  # lain jika 6 keluar dari looping
  elif menu == 6 :
    break
  # selain dari itu maka menu tidak valid
  else :
    print("Input yang di berikan salah !!!")
    continue
```

```python
# Membuat class AntrianFotokopi
class AntrianFotokopi :
  # Methpd Constructor
  def __init__(self) :
    # Atribut ini nantinya akan menjadi 2D seperti matriks
    # Yang mana pada dimensi ke 2 berisi [antrian, jumlahHalaman, Tahapan]
    self.antrian = []

  # Method untuk menampilkan pesanan pertama
  def front(self) :
    # Operator Kondisional
    # Jika datanya kosong maka mengembalika none jika ada datanya mengembalikan nilai antrean dengan indeks pertama
    return None if self.is_empty() else self.antrian[0]

  # Method untuk menampilkan pesan paling akhir
  def rear(self) :
    # Operator Kondisional
    # Jika datanya kosong maka mengembalika none jika ada datanya mengembalikan nilai antrean dengan indeks terakhir
    return None if self.is_empty() else self.antrian[-1]

  #   Method untuk cek apakah datanya kosong atau tidak
  def is_empty(self) :
    # Mengembalikan true jika datanya kosong
    return len(self.antrian) == 0

  # Method untuk mengembalikan panjang
  def size(self) :
    # Mengembalikan panjang dari nilai self.antrian
    return len(self.antrian)

  # Fitur yang memrposes halaman jika kurang dari 10 halaman
  def fast_track(self, num_pages) :
    return num_pages < 10;


   # Method untuk menambahkan pelanggan ke dalam antrian
  def enqueue(self, order_id, num_pages) :
    # Jika halamannya lebih dari 100 halaman
    if (num_pages > 100) :
      # Akan memasukan array yang berisi order_id, num_pages dan tahapan
      self.antrian.append([order_id, num_pages, 0])
      return
    # Jika dia halamannya kurang dari 10 maka akan di taruh di paling depan supaya di proses terlebih dahulu
    if self.fast_track(num_pages) :
      self.antrian.insert(0, [order_id, num_pages])
      print(f"Karena anda melakuan fotokopi dibawah 10 halaman, maka akan diproses terlebih dahulu")
      print("Memproses:", self.dequeue())
      return

    # Asumsi data tidak kurang dari 10 halaman dan lebih dari 100 halaman
    self.antrian.append([order_id, num_pages])
    return

  # Method untuk memproses data tersebut
  def dequeue(self) :
    # Jika datanya kosong
    if self.is_empty() : return None

    # Cek apakah antrian ini lebih dari 100 halaman atau tidak
    if len(self.antrian[0]) == 3 :
      # cek apakah sudah di proses 2 kali atau belum
      if self.antrian[0][2] < 1 :
        # Berarti belum melewati tahap apapun
        self.antrian[0][2] += 1
        print(f"{self.antrian[0][1]} sudah melewati tahap pertama")
        return
      # Jika sudah melewati tahap pertama
      elif self.antrian[0][2] == 1:
        return self.antrian.pop(0)

    # Asumsi datanya ada
    return self.antrian.pop(0)


antrianFotokopi = AntrianFotokopi()

while True :
  print("")
  print("Selamat Datang di Antrian Fotokopi")
  print(f"Antrian saat ini : {antrianFotokopi.antrian}")
  print("1. Menambahkan antrian.")
  print("2. Memproses antrian.")
  print("3. Kode antrian pertama.")
  print("4. Kode antrian terakhir.")
  print("5. Panjang antrian.")
  print("6. Keluar.")

  print("")
  # Input menu yang di pilih user
  menu = int(input("Pilihlah salah satu dari menu tersebut => "))
  print("")
  # Menu pilihan
  # Jika menu == 1 maka menambahkan kode antrian
  if menu == 1 :
    antrianFotokopi.enqueue(f"{str(input('Masukan kode antrian => '))}", int(input("Masukan jumlah halaman => ")))
  # lain jika 2 memproses kode antrian
  elif menu == 2:
    print("Memproses:", antrianFotokopi.dequeue())
  # lain jika 3 menampilkan kode antrian pertama
  elif menu == 3 :
    print("Kode antrian pertama:", antrianFotokopi.front())
  # lain jika 4 menampilkan kode antrian terakhir
  elif menu == 4 :
    print("Rear element:", antrianFotokopi.rear())
  # lain jika 5 menampilkan panjang antrian
  elif menu == 5 :
    print("Panjang antrian:", antrianFotokopi.size())
  # lain jika 6 keluar dari looping
  elif menu == 6 :
    break
  # selain dari itu maka menu tidak valid
  else :
    print("Input yang di berikan salah !!!")
    continue
```

```python
# Membuat class Queue dengan nama AntreanParkir
class AntreanParkir :
  # Constructor Method
  def __init__(self) :
    self.antrian = []

  # Membuat method yang mana parkir prioritas memiliki prioritas masuk parkiran lebih dahulu
  def parkirPrioritas(self, vehicle_number) :
    print("Parkir Prioritas")
    self.antrian.insert(0, vehicle_number)
    print(f"Plat antrian {self.dequeue()} berhasil di proses")
    return 0

  # Method untuk menambahkan mobil ke dalam antrian parkir
  def enqueue(self, vehicle_number) :
    # Jika antriannya penuh maka harus menunggu hinga slot kosong batas antrian 15
    if (self.size() >= 15) :
      print("Kapasitas antrian parkir penuh, tunggu hingga kapasitas kosong.");
      return 0
    # Menambahkan vehicle_number ke self.antrian
    self.antrian.append(vehicle_number)
    # Menampilkan pesan bahwa antrian berhasil ditambahkan
    print(f"Plat kendaraan : {vehicle_number} berhasil di tambahkan kedalam antrian")

  # Method untuk mengembalikan panjang dari antrian
  def size(self) :
    # Mengembalikan panjang dari nilai self.antrian
    return len(self.antrian)

  # Memproses antrian dan mengeluarkannya dari antrian
  def dequeue(self) :
    # Jika datanya kosong
    if self.is_empty() : return None

    # Asumsi datanya ada
    return self.antrian.pop(0)

  # Method untuk cek apakah datanya kosong atau tidak
  def is_empty(self) :
    # Mengembalikan true jika datanya kosong
    return len(self.antrian) == 0

  # Method untuk menampilkan elemen paling depan
  def front(self) :
    # Operator Kondisional
    # Jika datanya kosong maka mengembalika none jika ada datanya mengembalikan nilai antrean dengan indeks pertama
    return None if self.is_empty() else self.antrian[0]

  # Method untuk menampilkan data paling akhir
  def rear(self) :
    # Operator Kondisional
    # Jika datanya kosong maka mengembalika none jika ada datanya mengembalikan nilai antrean dengan indeks terakhir
    return None if self.is_empty() else self.antrian[-1]

# Instansi class AntreanParkir
antrian = AntreanParkir()


# Tes 15 Kendaraan
for i in range(0,15) :
  antrian.enqueue(f"BE{i}")

while True :
  print("")
  print("Selamat Datang di Antrian Parkir Mall")
  print(f"Antrian parkir saat ini : {antrian.antrian}")
  print("1. Menambahkan antrian.")
  print("2. Memproses antrian.")
  print("3. Plat antrian pertama.")
  print("4. Plat antrian terakhir.")
  print("5. Panjang antrian.")
  print("6. Parkiran Prioritas.")
  print("7. Keluar.")

  print("")
  # Input menu yang di pilih user
  menu = int(input("Pilihlah salah satu dari menu tersebut => "))
  print("")
  # Menu pilihan
  # Jika menu == 1 maka menambahkan Plat antrian
  if menu == 1 :
    antrian.enqueue(f"{str(input('Masukan Plat antrian => '))}")
  # lain jika 2 memproses Plat antrian
  elif menu == 2:
    print("Memproses:", antrian.dequeue())
  # lain jika 3 menampilkan Plat antrian pertama
  elif menu == 3 :
    print("Kendaraan antrian pertama:", antrian.front())
  # lain jika 4 menampilkan Plat antrian terakhir
  elif menu == 4 :
    print("Kendaraan antrian terakhir:", antrian.rear())
  # lain jika 5 menampilkan panjang antrian
  elif menu == 5 :
    print("Panjang antrian:", antrian.size())
  # lain jika 6 parkirPrioritas
  elif menu == 6 :
    antrian.parkirPrioritas(f"{str(input('Masukan Plat antrian => '))}")
  elif menu == 7 :
    break
  # selain dari itu maka menu tidak valid
  else :
    print("Input yang di berikan salah !!!")
    continue
```

```python
# Membuat class Queue dengan nama PengambilanSertifikat
class PengambilanSertifikat :
  # Constructor Method
  def __init__(self) :
    self.antrian = []
    # atribut untuk dua jalur
    self.jalur = False

  # Membuat method yang mana mahasiswa berprestasi memiliki prioritas menggambil sertifikat terlebih dahulu
  def mahasiswaBerprestasi(self, student_id) :
    # Jika jalur belum dibagi 2
    if self.jalur == False :
      print("Mahasiswa Berprestasi")
      self.antrian.insert(0, student_id)
      print(f"Nommor ID {self.dequeue()} berhasil di proses")
      return 0
    else :
      print("Mahasiswa Berprestasi")
      self.antrian["Prestasi"].insert(0, student_id)
      print(f"Nommor ID {self.dequeue()} berhasil di proses")

  # Method untuk menambahkan mahaiswa ke dalam antrian pengambilan sertifikat
  def enqueue(self, student_id) :
    if (self.jalur == False) :
      # Jika antriannya penuh maka harus membagi dua jalur batas antrian 15
      if (self.size() >= 20) :
        print("Antrian pengambilan sertifikat penuh, membagi menjadi 2 jalur.");
        # Spliting data antrian
        self.antrian = {"Reguler" : self.antrian[:9],
                    "Prestasi" : self.antrian[9:]}
        self.jalur = True
        return 0
      # Menambahkan student_id ke self.antrian
      self.antrian.append(student_id)
      # Menampilkan pesan bahwa antrian berhasil ditambahkan
      print(f"Nomor ID : {student_id} berhasil di tambahkan kedalam antrian")
    # Jika Jalur dibagi menjadi 2 maka
    elif self.jalur == True :
      # Cek mahasiswa jalur reguler atau tidak
      if (student_id[0] == "reguler") :
        # Jika ya tambahkan mahasiswa reguler
        self.antrian['Reguler'].append(student_id[1])
      else :
        # Jika tidak tambahkan mahasiswa berprestasi
        self.antrian['Prestasi'].append(student_id[1])



  # Method untuk mengembalikan panjang dari antrian
  def size(self) :
    # Mengembalikan panjang dari nilai self.antrian
    return len(self.antrian)

  # Memproses antrian dan mengeluarkannya dari antrian
  def dequeue(self) :
    # Jika datanya kosong
    if self.is_empty() : return None

    # Jika jalur belum dibagi 2
    if self.jalur == False :
      # Asumsi datanya ada
      return self.antrian.pop(0)
    # Jika jalurnya sudah dibagi 2
    elif self.jalur == True :
      # akan memproses kedua jalur agar proses bisa lebih cepat
      a = {"Reguler" : self.antrian["Reguler"].pop(),
          "Prestasi" : self.antrian["Prestasi"].pop()}


      # Jika salah satu jalur ada yang kosong maka jalur akan kembali seperti semula
      if (len(self.antrian["Reguler"]) == 0 or len(self.antrian["Prestasi"]) == 0) :
        # Kembalikan jalur seperti semula
        self.jalur = False

        # Simpan reguler dan prestasi ke variabel sementara
        reguler = self.antrian["Reguler"]
        prestasi = self.antrian["Prestasi"]

        # Kosongkan antrian
        self.antrian = []

        # Lalu masukan satu persatu
        self.antrian.extend(reguler)
        self.antrian.extend(prestasi)
      return a

  # Method untuk cek apakah datanya kosong atau tidak
  def is_empty(self) :
    # Mengembalikan true jika datanya kosong
    return len(self.antrian) == 0

  # Method untuk menampilkan elemen paling depan
  def front(self) :
    # Operator Kondisional
    # Jika datanya kosong maka mengembalika none jika ada datanya mengembalikan nilai antrean dengan indeks pertama
    return None if self.is_empty() else self.antrian[0] if self.jalur == False else {"Reguler" : self.antrian["Reguler"][0], "Prestasi" : self.antrian["Prestasi"][0]}

  # Method untuk menampilkan data paling akhir
  def rear(self) :
    # Operator Kondisional
    # Jika datanya kosong maka mengembalika none jika ada datanya mengembalikan nilai antrean dengan indeks terakhir
    return None if self.is_empty() else self.antrian[-1] if self.jalur == False else {"Reguler" : self.antrian["Reguler"][-1], "Prestasi" : self.antrian["Prestasi"][-1]}

# Instansi class PengambilanSertifikat
antrian = PengambilanSertifikat()


# Tes 20 mahasiswa
for i in range(0,20) :
  antrian.enqueue(f"T{i}")

while True :
  print("")
  print("Selamat Datang di Antrian Pengambilan sertifikat")
  print(f"Antrian saat ini : {antrian.antrian}")
  print("1. Menambahkan antrian.")
  print("2. Memproses antrian.")
  print("3. Mahasiswa ID antrian pertama.")
  print("4. Mahasiswa ID antrian terakhir.")
  print("5. Panjang antrian.")
  print("6. Mahasiswa Berprestasi.")
  print("7. Keluar.")

  print("")
  # Input menu yang di pilih user
  menu = int(input("Pilihlah salah satu dari menu tersebut => "))
  print("")
  # Menu pilihan
  # Jika menu == 1 maka menambahkan Mahasiswa ID antrian
  if menu == 1 :
    if antrian.jalur == False :
      antrian.enqueue(f"{str(input('Masukan Mahasiswa ID antrian => '))}")
    else :
      antrian.enqueue([f"{str(input('Mahasiswa (Reguler / Prestasi) => '))}".lower(),f"{str(input('Masukan Mahasiswa ID antrian => '))}"])
  # lain jika 2 memproses Mahasiswa ID antrian
  elif menu == 2:
    print("Memproses:", antrian.dequeue())
  # lain jika 3 menampilkan Mahasiswa ID antrian pertama
  elif menu == 3 :
    print("Mahasiswa antrian pertama:", antrian.front())
  # lain jika 4 menampilkan Mahasiswa ID antrian terakhir
  elif menu == 4 :
    print("Mahasiswa antrian terakhir:", antrian.rear())
  # lain jika 5 menampilkan panjang antrian
  elif menu == 5 :
    print("Panjang antrian:", antrian.size())
  # lain jika 6 mahasiswaBerprestasi
  elif menu == 6 :
    antrian.mahasiswaBerprestasi(f"{str(input('Masukan Mahasiswa ID antrian => '))}")
  elif menu == 7 :
    break
  # selain dari itu maka menu tidak valid
  else :
    print("Input yang di berikan salah !!!")
    continue
```
